# API e Dados em Python

## Prática com Gemini, laços, listas e dicionários

Neste notebook vou praticar:
- conexão com a API do Gemini
- envio de prompts
- criação de chat interativo
- uso de while
- listas
- dicionários
- manipulação de dados em Python

```python
# Configuração inicial da API do Gemini no Google Colab

import os
from google.colab import userdata
```

```python
# Recuperando a chave salva no cofre Secrets do Colab

os.environ["GOOGLE_API_KEY"] = userdata.get("python-ai-integrada")

print("Chave carregada com sucesso!")
```

```python
# Importando o SDK do Google Gen AI
from google import genai

# Criando o cliente
client = genai.Client()

print("Cliente Gemini criado com sucesso!")
```

```python
# Enviando um prompt simples para o Gemini

resposta = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="O que é a inteligência artificial?"
)

print("Resposta da IA:\n")
print(resposta.text)
```

```python
# Enviando um prompt simples para o Gemini com exceção

try:
    resposta = client.models.generate_content(
        model="gemini-2.5-flash",
        contents="O que é a inteligência artificial?"
    )
    print(resposta.text)
except Exception as e:
    print("Ocorreu um erro ao consultar o Gemini:")
    print(e)
```

```python
# Exemplo simples de estrutura de repetição com while

contador = 0

while contador < 5:
    print("*" * (contador + 1))
    contador += 1
```

```python
# Capturando dados do usuário

nome = input("Digite seu nome: ")
print(f"Olá, {nome}! Vamos testar um chat com IA.")
```

```python
# Criando um chat com contexto

chat = client.chats.create(model="gemini-2.5-flash")

print("Chat iniciado com sucesso!")
```

```python
# Chat interativo com condição de parada

prompt = input("Digite sua pergunta para a IA (ou 'fim' para encerrar): ")

while prompt.lower() != "fim":
    resposta = chat.send_message(prompt)

    print("\nResposta da IA:")
    print(resposta.text)
    print("\n" + "-" * 50 + "\n")

    prompt = input("Digite outra pergunta (ou 'fim' para encerrar): ")

print("Chat encerrado.")
```

```python
# Exibindo o histórico completo do chat

historico = chat.get_history()

print("Histórico das interações:\n")

for item in historico:
    print(item)
    print("-" * 50)
```

## Trabalhando com listas

Agora vamos praticar listas, índices, slices e atualização de valores.

```python
# Criando listas de nomes e médias

nomes = ["Ana", "Bruno", "Carlos", "Daniela"]
medias = [7.5, 8.0, 6.5, 9.0]

print("Lista de nomes:", nomes)
print("Lista de médias:", medias)
```

```python
# Acessando elementos específicos

print("Primeiro aluno:", nomes[0])
print("Segunda média:", medias[1])
print("Último aluno:", nomes[-1])
```

```python
# Utilizando slices

print("Dois primeiros nomes:", nomes[:2])
print("Últimas duas médias:", medias[-2:])
```

```python
# Alterando um valor da lista

medias[2] = 7.0

print("Lista de médias atualizada:", medias)
```

```python
# Percorrendo os elementos da lista

for nome in nomes:
    print(nome)
```

```python
# Aumentando 1 ponto na média, sem ultrapassar 10

for i in range(len(medias)):
    medias[i] += 1

    if medias[i] > 10:
        medias[i] = 10

print("Médias após incremento:", medias)
```

```python
# Criando dicionários com nome e média dos alunos

alunos = []

for i in range(len(nomes)):
    aluno = {
        "nome": nomes[i],
        "media": medias[i]
    }
    alunos.append(aluno)

print("Lista de alunos com dicionários:")
print(alunos)
```

```python
# Exibindo de forma mais legível

for aluno in alunos:
    print(f"Nome: {aluno['nome']} | Média: {aluno['media']}")
```

```python
# Resumo final da atividade

print("===== RESUMO DA ATIVIDADE =====")
print(f"Quantidade de alunos cadastrados: {len(alunos)}")

for aluno in alunos:
    if aluno["media"] >= 7:
        situacao = "Aprovado"
    elif aluno["media"] >= 5:
        situacao = "Recuperação"
    else:
        situacao = "Reprovado"

    print(f"{aluno['nome']} - Média: {aluno['media']} - Situação: {situacao}")
```
