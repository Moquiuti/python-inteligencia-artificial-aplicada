```python
cafe = 8.50
pao_de_queijo = 6.00
bolo = 12.00

total = cafe + pao_de_queijo + bolo
print("Total da compra:", total)

desconto = 5
valor_final = total - desconto
print("Valor final com desconto:", valor_final)

dobro_do_cafe = cafe * 2
print("Dois cafés custam:", dobro_do_cafe)

divisao_conta = valor_final / 2
print("Dividindo a conta por 2 pessoas:", divisao_conta)
```

```python
quantidade_pessoas = 3
valor_por_pessoa = valor_final / quantidade_pessoas

print("Valor total:", valor_final)
print("Quantidade de pessoas:", quantidade_pessoas)
print("Cada pessoa pagará:", valor_por_pessoa)
```

```python
nome = input("Digite seu nome: ").strip().title()
pedido = input("Digite o produto desejado: ").strip().lower()
quantidade = int(input("Digite a quantidade: "))
idade = int(input("Digite sua idade: "))

pedido = pedido.replace("cafe", "café")

preco_unitario = 10.0
subtotal = quantidade * preco_unitario

if quantidade >= 3:
    desconto = 5
else:
    desconto = 0

total = subtotal - desconto

print("\n===== RESUMO DO PEDIDO =====")
print(f"Cliente: {nome}")
print(f"Produto: {pedido}")
print(f"Quantidade: {quantidade}")
print(f"Subtotal: R$ {subtotal:.2f}")
print(f"Desconto: R$ {desconto:.2f}")
print(f"Total final: R$ {total:.2f}")

if idade >= 18:
    print("Cliente autorizado para qualquer item do cardápio.")
else:
    print("Cliente menor de idade.")
```

```python
nome = input("Digite o nome do aluno: ").strip().title()
idade = int(input("Digite a idade do aluno: "))
nota = float(input("Digite a nota do aluno: "))

print("\n===== RESULTADO =====")
print(f"Aluno: {nome}")
print(f"Idade: {idade}")
print(f"Nota: {nota}")

if idade >= 18:
    print("Situação de idade: maior de idade")
else:
    print("Situação de idade: menor de idade")

if nota >= 7:
    print("Resultado escolar: aprovado")
elif nota >= 5:
    print("Resultado escolar: recuperação")
else:
    print("Resultado escolar: reprovado")
```
