```python
# Lista de e-mails de exemplo

emails = [
    "Olá equipe, precisamos revisar o cronograma do projeto e validar os prazos com o fornecedor.",
    "Cliente relatou dificuldade no acesso ao portal após redefinir a senha.",
    "Concluímos a reunião de alinhamento e vamos priorizar as integrações críticas até sexta-feira."
]
```

```python
# Função para gerar resumos simples sem LLM, apenas para praticar lista, append e f-string

def gerar_resumos(lista_emails):
    lista_de_resumos = []

    for indice, email in enumerate(lista_emails, start=1):
        resumo = f"E-mail {indice}: {email[:60]}..."
        lista_de_resumos.append(resumo)

    return lista_de_resumos
```

```python
resumos = gerar_resumos(emails)

for item in resumos:
    print(item)
```

```python
# Salvando os resumos em arquivo txt

with open("resumos_emails.txt", "w", encoding="utf-8") as arquivo:
    for resumo in resumos:
        arquivo.write(resumo + "\n")
```

```python
# Lendo linha por linha e salvando em nova lista, removendo quebra de linha

linhas_lidas = []

with open("resumos_emails.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        linhas_lidas.append(linha.strip())

linhas_lidas
```

```python
# read() -> lê tudo de uma vez

with open("resumos_emails.txt", "r", encoding="utf-8") as arquivo:
    conteudo_completo = arquivo.read()

print("Usando read():\n")
print(conteudo_completo)
```

```python
# readline() -> lê uma linha por vez

with open("resumos_emails.txt", "r", encoding="utf-8") as arquivo:
    primeira_linha = arquivo.readline().strip()
    segunda_linha = arquivo.readline().strip()

print("Primeira linha:", primeira_linha)
print("Segunda linha:", segunda_linha)
```

```python
# readlines() -> devolve lista de linhas

with open("resumos_emails.txt", "r", encoding="utf-8") as arquivo:
    todas_as_linhas = arquivo.readlines()

print(todas_as_linhas)
```

```python
# Limpando as quebras de linha vindas do readlines()

linhas_tratadas = [linha.strip() for linha in todas_as_linhas]
print(linhas_tratadas)
```

```python
perguntas = [
    "O que é inteligência artificial?",
    "O que é um DataFrame no pandas?",
    "Qual a diferença entre read e readlines em Python?",
    "Para que serve um arquivo CSV?"
]
```

```python
with open("perguntas.txt", "w", encoding="utf-8") as arquivo:
    for pergunta in perguntas:
        arquivo.write(pergunta + "\n")
```

```python
lista_perguntas_lidas = []

with open("perguntas.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        lista_perguntas_lidas.append(linha.strip())

lista_perguntas_lidas
```

```python
import os
from google.colab import userdata
from groq import Groq

os.environ["GROQ_API_KEY"] = userdata.get("CursoAluraPytonIaGenerative")
client_groq = Groq()

print("Cliente Groq configurado com sucesso.")
```

```python
def responder_pergunta(pergunta):
    resposta = client_groq.chat.completions.create(
        model="openai/gpt-oss-20b",
        messages=[
            {
                "role": "system",
                "content": (
                    "Responda em português de forma curta, clara e objetiva, em no máximo 2 frases."
                )
            },
            {
                "role": "user",
                "content": pergunta
            }
        ],
        temperature=0.2,
        reasoning_effort="low",
        max_completion_tokens=300
    )

    conteudo = resposta.choices[0].message.content

    if conteudo is None:
        return "Resposta não retornada pela API."

    conteudo = conteudo.strip()

    if conteudo == "":
        return "Resposta veio vazia."

    return conteudo
```

```python
resultado_final = []

for pergunta in lista_perguntas_lidas:
    resposta = responder_pergunta(pergunta)

    resultado_final.append({
        "pergunta": pergunta,
        "resposta": resposta
    })

resultado_final
```

```python
df_resultado = pd.DataFrame(resultado_final)
df_resultado
```

```python
df_resultado.to_csv("perguntas_e_respostas.csv", index=False, encoding="utf-8")
print("Arquivo final salvo com sucesso.")
```
