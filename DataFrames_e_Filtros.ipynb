```python
import os
from google.colab import userdata
from google import genai
```

```python
os.environ["GOOGLE_API_KEY"] = userdata.get("python-ai-integrada")
client = genai.Client()
```

```python
df_reviews = pd.read_csv("reviews.csv")
display(df_reviews.head())
```

```python
print(df_reviews.columns)
```

```python
import time

def classificar_sentimento(review):
    prompt = f"""
Você é um classificador de sentimentos.

Classifique a resenha abaixo retornando apenas uma destas opções:
Positiva
Negativa
Neutra

Exemplos:
Resenha: "Produto excelente, superou minhas expectativas."
Classificação: Positiva

Resenha: "Não gostei, veio com defeito e parou de funcionar."
Classificação: Negativa

Resenha: "O produto é ok, nada demais."
Classificação: Neutra

Agora classifique:
Resenha: "{review}"
Classificação:
"""

    for tentativa in range(3):
        try:
            resposta = client.models.generate_content(
                model="gemini-2.5-flash",
                contents=prompt
            )

            resultado = resposta.text.strip()

            if "Positiva" in resultado:
                return "Positiva"
            elif "Negativa" in resultado:
                return "Negativa"
            elif "Neutra" in resultado:
                return "Neutra"
            else:
                return "Neutra"

        except Exception as e:
            print(f"Tentativa {tentativa + 1} falhou: {e}")
            time.sleep(2)

    return "Neutra"
```

```python
lista_de_analises_de_sentimentos = []

for review in df_reviews["reviewText"]:
    sentimento = classificar_sentimento(review)
    lista_de_analises_de_sentimentos.append(sentimento)
```

```python
df_reviews["Análises de Sentimentos"] = lista_de_analises_de_sentimentos
display(df_reviews.head())
```

```python
df_reviews.to_csv("reviews_com_sentimento.csv", index=False, encoding="utf-8")
print("Arquivo reviews_com_sentimento.csv salvo com sucesso.")
```
