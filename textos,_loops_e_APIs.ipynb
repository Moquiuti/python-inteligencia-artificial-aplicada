```python
def corrigir_texto(texto):
    texto = texto.strip()
    texto = texto.upper()
    texto = texto.replace("  ", " ")

    # elimina espaços extras de forma definitiva
    palavras = texto.split()
    texto = " ".join(palavras)

    return texto
```

```python
print("Nomes corrigidos:\n")

for nome in nomes_brutos:
    nome_corrigido = corrigir_texto(nome)
    print(nome_corrigido)
```

```python
def corrigir_nome_proprio(nome):
    nome = nome.strip()
    palavras = nome.split()
    palavras_corrigidas = []

    for palavra in palavras:
        palavras_corrigidas.append(palavra.capitalize())

    return " ".join(palavras_corrigidas)
```

```python
print("Padronização em formato de nome próprio:\n")

nomes_corrigidos = []

for nome in nomes_brutos:
    nome_corrigido = corrigir_nome_proprio(nome)
    nomes_corrigidos.append(nome_corrigido)
    print(nome_corrigido)
```

```python
import random

setores = ["Financeiro", "Comercial", "Suporte", "Operações"]
status_possiveis = ["Novo", "Em análise", "Respondido"]

cadastro_emails = []

for nome in nomes_corrigidos:
    registro = {
        "nome": nome,
        "setor": random.choice(setores),
        "status": random.choice(status_possiveis)
    }
    cadastro_emails.append(registro)

cadastro_emails
```

```python
print("Exibindo cada dicionário:\n")

for item in cadastro_emails:
    print(item)
```

```python
print("Percorrendo dicionário com chaves e valores:\n")

for item in cadastro_emails:
    print(f"Dados de {item['nome']}:")

    for chave, valor in item.items():
        print(f"  {chave} -> {valor}")

    print("-" * 40)
```

```python
def limpar_email(texto):
    texto = texto.strip()
    palavras = texto.split()
    texto = " ".join(palavras)
    return texto
```

```python
print("E-mails padronizados:\n")

for i, email in enumerate(emails, start=1):
    email_limpo = limpar_email(email)
    print(f"E-mail {i}: {email_limpo}")
    print("-" * 80)
```

```python
def resumir_email(email_texto):
    email_limpo = limpar_email(email_texto)

    resposta = client.chat.completions.create(
        model="openai/gpt-oss-20b",
        messages=[
            {
                "role": "system",
                "content": (
                    "Você é um assistente que resume e-mails corporativos em português. "
                    "Resuma com objetividade em no máximo 2 frases, destacando o ponto principal."
                )
            },
            {
                "role": "user",
                "content": f"Resuma o seguinte e-mail:\n\n{email_limpo}"
            }
        ],
        temperature=0.3,
        max_completion_tokens=120
    )

    return resposta.choices[0].message.content
```

```python
teste_resumo = resumir_email(emails[0])

print("Resumo de teste:\n")
print(teste_resumo)
```

```python
def resumir_lista_emails(lista_emails):
    resumos = []

    for indice, email in enumerate(lista_emails, start=1):
        resumo = resumir_email(email)

        item = {
            "numero": indice,
            "email_original": limpar_email(email),
            "resumo": resumo
        }

        resumos.append(item)

    return resumos
```

```python
resumos_gerados = resumir_lista_emails(emails)
resumos_gerados
```

```python
print("Resumos gerados pela IA:\n")

for item in resumos_gerados:
    print(f"E-mail {item['numero']}")
    print(f"Conteúdo original: {item['email_original']}")
    print(f"Resumo: {item['resumo']}")
    print("=" * 100)
```

```python
caixa_entrada = []

for i in range(len(emails)):
    remetente = nomes_corrigidos[i % len(nomes_corrigidos)]

    caixa_entrada.append({
        "remetente": remetente,
        "setor": random.choice(setores),
        "status": random.choice(status_possiveis),
        "email": limpar_email(emails[i])
    })

caixa_entrada
```

```python
def resumir_caixa_entrada(caixa):
    saida = []

    for indice, item in enumerate(caixa, start=1):
        resumo = resumir_email(item["email"])

        registro = {
            "id": indice,
            "remetente": item["remetente"],
            "setor": item["setor"],
            "status": item["status"],
            "resumo": resumo
        }

        saida.append(registro)

    return saida
```

```python
caixa_resumida = resumir_caixa_entrada(caixa_entrada)

for item in caixa_resumida:
    print(f"Mensagem {item['id']}")
    print(f"Remetente: {item['remetente']}")
    print(f"Setor: {item['setor']}")
    print(f"Status: {item['status']}")
    print(f"Resumo: {item['resumo']}")
    print("-" * 100)
```

```python
def resumir_email(email_texto):
    email_limpo = limpar_email(email_texto)

    resposta = client.chat.completions.create(
        model="openai/gpt-oss-20b",
        messages=[
            {
                "role": "system",
                "content": (
                    "Você é um assistente que resume e-mails corporativos em português. "
                    "Resuma com objetividade em no máximo 2 frases."
                )
            },
            {
                "role": "user",
                "content": f"Resuma o seguinte e-mail:\n\n{email_limpo}"
            }
        ],
        temperature=0.3,
        max_completion_tokens=120
    )

    conteudo = resposta.choices[0].message.content

    if conteudo is None:
        return "Resumo não retornado pela API."

    conteudo = conteudo.strip()

    if conteudo == "":
        return "Resumo veio vazio."

    return conteudo
```

```python
def resumir_email_com_debug(email_texto):
    email_limpo = limpar_email(email_texto)

    resposta = client.chat.completions.create(
        model="openai/gpt-oss-20b",
        messages=[
            {
                "role": "system",
                "content": (
                    "Você resume e-mails corporativos em português. "
                    "Responda apenas com um resumo curto e objetivo."
                )
            },
            {
                "role": "user",
                "content": f"Resuma este e-mail:\n\n{email_limpo}"
            }
        ],
        temperature=0.2,
        reasoning_effort="low",
        max_completion_tokens=300
    )

    print("Finish reason:", resposta.choices[0].finish_reason)
    print("Usage:", resposta.usage)

    conteudo = resposta.choices[0].message.content

    if conteudo is None:
        return "Resumo não retornado pela API."

    conteudo = conteudo.strip()

    if conteudo == "":
        return "Resumo veio vazio."

    return conteudo
```
